```python
import pycountry
```

```python
country = []
for i in pycountry.countries:
    country.append(i.name)
```

```python
for i in range(len(country)):
    country[i] = country[i].lower()
```

```python
country.append('america')
country.append('uk')
country.append('usa')
country.append('usa')
country.append('sg')
country.append('usa')
```

```python
def extract_countries(text):
    tokens = []
    for word in text.split():
        if word in country:
            tokens.append(word)
    return ", ".join(tokens)
```

```python
data['countries'] = data['Tweet'].apply(lambda x: extract_countries(x))
```
